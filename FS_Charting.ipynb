```python
import pandas as pd

import matplotlib.pyplot as plt

file_path = '/Users/ahmedshaker/Downloads/Balance Sheet Over Period.xlsx'

# Read the Excel file and create a DataFrame
df = pd.read_excel(file_path, engine='openpyxl')

# Check if the 'Month' column exists in your DataFrame
if 'Month' not in df.columns:
    print("The 'Month' column was not found in the DataFrame. Please check the input data.")
    exit()

# Format the date values in the 'Month' column with month name and year
df['Month'] = df['Month'].dt.strftime('%b %Y')  # You can use '%B %Y' for the full month name

# Set the Month column as the index
df.set_index('Month', inplace=True)

# Create a vertical bar chart with bars side by side
ax = df.plot(kind='bar', figsize=(25, 5), width=.8)

# Set the x-tick labels at every 3rd month
ax.set_xticks(range(0, len(df.index), 3))

# Set the x-tick labels to be rotated at 45 degrees
ax.set_xticklabels(df.index[::3], rotation=45)

ax.set_yticks(range(0, 7500001, 2500000))
ax.set_yticklabels([f'{y:,.0f} $' for y in ax.get_yticks()])

# Add labels and titles
ax.set_xlabel('Month')
ax.set_ylabel('Amount')
ax.set_title('Income Statement Westcore')

# Display the plot
plt.show()

```
